```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dados_pizza.csv')
df.head()
```

```python

plt.figure(figsize = (12,5))
sns.set(font_scale=1)

sns.scatterplot(x='Valor (R$)',y='Diâmetro (cm)',data=df)
```

```python
# Separando os valores das colunas e transformando-os em uma matriz unidimensional
valor = df['Valor (R$)'].values.reshape(-1,1)

diametro = df['Diâmetro (cm)'].values.reshape(-1,1)
```

```python
from sklearn import linear_model
regressor = linear_model.LinearRegression()
regressor.fit(diametro,valor)

```

```python
from sklearn.model_selection import cross_val_predict
previsao = cross_val_predict(regressor, diametro, valor, cv=5)
```

```python
plt.figure(figsize=(12, 5))
sns.set(font_scale=1)
sns.scatterplot(x='Diâmetro (cm)', y='Valor (R$)', data=df, label='Dados Originais',alpha=0.8)
plt.plot(diametro, previsao, color='red', linewidth=1, label='Regressão Linear',linestyle=':')
plt.xlabel("Diâmetro (cm)")
plt.ylabel("Valor (R$)")
plt.title("Dados Originais vs. Regressão Linear")
plt.legend()
```
